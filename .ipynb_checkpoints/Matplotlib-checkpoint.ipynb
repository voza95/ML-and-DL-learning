```python
import numpy as np
```

```python
t = np.arange(0., 5., 0.2)
# Blue dashes, Red squares, and Green triangles
plt.plot(t, t**2, 'b--', label='^2', linewidth=5)
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

# Working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
# The subplot command specifies numrows(2), numcols(1), figure(1) where fignum ranges from 1 to numrow*numcols
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')

# numrows(2), numcols(1), figure(2)
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
